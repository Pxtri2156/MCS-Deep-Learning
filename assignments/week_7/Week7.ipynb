# Exercise 3
* Input Image: 7x7
* Kernel: 3x3
* Output: 5x5

Hint: For each dimension, N=image_width, m kernel (mask) width, s=step size, output_feature_map_width={[(N-m)/s]+1}

## a) If the step size of the correlation is 1 pixel (horizontally and vertically), explain why the above output feature map is 5x5.
=> **Answer**:  `output_feature_map_width={[(N-m)/s]+1} = {[(7-3)/1]+1} = 5`

## b) If input is 32x32, mask is 5x5, what is the size of the output feature map?
=> **Answer**:  `output_feature_map_width={[(N-m)/s]+1} = {[(32-5)/1]+1} = 28`

## c) If input is 28x28, what is the size of the subsample layer?
=> **Answer**:  `output_feature_map_width={[(N-m)/s]+1} = {[(28-3)/1]+1} = 26`

## d) If input is 14x14, kernel-5x5, what is the size of the output feature map?
=> **Answer**:  `output_feature_map_width={[(N-m)/s]+1} = {[(14-5)/1]+1} = 10`

## e) In question(a), if the step size of the convolution is 2 pixels, What is the size of he output feature map
=> **Answer**:  `output_feature_map_width={[(N-m)/s]+1} = {[(7-3)/2]+1} = 3`
