```python
import pandas as pd
from pyvi import ViTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import preprocessing
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from bs4 import BeautifulSoup
import requests
from sklearn.neighbors import KNeighborsClassifier
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
url_data = pd.read_excel('/content/drive/MyDrive/MCS/Deep Learning/Week 1/Homework1/Homework1_DATA.xlsx')
print(url_data)
```

```python
X_raw = []
Y_raw = []
for _, row in url_data.iterrows():
    url = row['url']
    print(url)
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content, "lxml")
    # data = soup.find("section", attrs={"class": "section page-detail top-detail"}).find_all('p', attrs={"class": "Normal"})
    data = soup.find_all('p', attrs={"class": "Normal"})
    content = ""
    for element in data:
      # print(element.get_text())
      content += element.get_text()
    # print(content)

    #raw data
    X_raw.append(content)
    Y_raw.append(row['category'])

    #Tokenize
    content_token = ViTokenizer.tokenize(content).split(' ')
    # print(len(content_token))
```

```python
print(len(X_raw), len(Y_raw))
print(X_raw)
print(Y_raw)
```

```python
#IF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X_raw)
Y = np.array(Y_raw)

#number category 
le = preprocessing.LabelEncoder()
le.fit(Y)
Y = le.transform(Y)
print(X.shape)
print(Y)
```

```python
#Đánh giá trên KNN với kfold = 5 
kfold = model_selection.KFold(n_splits=5)
cv_results = model_selection.cross_val_score(KNeighborsClassifier(), X, Y, cv=kfold, scoring='accuracy')
print(cv_results.mean())
```

```python
#split dataset
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
print(X_train.shape)
print(X_train)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
bayes = KNeighborsClassifier()
bayes.fit(X_train, y_test)
predictions = bayes.predict(X_test)
print(accuracy_score(y_test, predictions))
print("\t=============================================\n")
print(confusion_matrix(y_test, predictions))
print("\t=============================================\n")
print(f1_score(y_test, predictions))
```

## **Câu 1 Dựa trên bộ dữ liệu tự thu thập từ các trang báo mạng Tiếng Việt (5 lớp/ mỗi lớp 10 mẩu tin), hãy thực hiện các yêu cầu sau:**

Kiến trúc tổng quát của ChatGPT dựa trên mô hình GPT-3.5, một mô hình ngôn ngữ tổng quát dựa trên học sâu. Mô hình này sử dụng một kiến trúc mạng nơ-ron tái cổ điển gọi là Transformer, được đề xuất bởi Vaswani et al. (2017).

Kiến trúc GPT-3.5 bao gồm các thành phần chính sau:

1. Đầu vào mã hóa: Văn bản đầu vào được mã hóa thành dạng vector trước khi đưa vào 
mô hình. Mã hóa có thể được thực hiện bằng cách sử dụng mã hóa vị trí (position encoding) và mã hóa từ (word encoding) để biểu diễn thông tin về vị trí và từ trong câu.

2. Mô hình Transformer: Mô hình Transformer là một mạng nơ-ron tái cổ điển được sử dụng để xử lý thông tin ngôn ngữ. Nó bao gồm một số lượng lớp tự-attention (self-attention) và lớp mạng nơ-ron truyền thẳng (feed-forward neural networks). Mỗi lớp tự-attention trong mô hình Transformer nhận đầu vào là một chuỗi các vector và tạo ra đầu ra tương tự với kích thước và cấu trúc giống nhau. Lớp mạng nơ-ron truyền thẳng sau đó được áp dụng lên mỗi vector trong chuỗi đầu ra của lớp tự-attention. Mô hình Transformer cũng sử dụng các cơ chế như Multi-Head Attention và Layer Normalization để cải thiện khả năng học và xử lý thông tin ngôn ngữ.

3. Giải mã và dự đoán: Sau khi thông tin đã được xử lý qua mô hình Transformer, các vector đầu ra được giải mã để tạo ra câu trả lời hoặc dự đoán. Trong trường hợp ChatGPT, nó được sử dụng để sinh ra phản hồi tự nhiên cho các câu hỏi hoặc yêu cầu từ người dùng.

4. Học không giám sát: GPT-3.5 cũng sử dụng một cơ chế học không giám sát gọi là Pretraining và Fine-tuning. Trong quá trình Pretraining, mô hình được huấn luyện trước trên một lượng lớn dữ liệu văn bản không có nhãn từ Internet. Quá trình Fine-tuning sau đó tiếp tục huấn luyện mô hình trên một tập dữ liệu có nhãn nhỏ hơn, được chỉ định cho nhiệm vụ cụ thể như hội thoại.  

Tóm lại, kiến trúc tổng quát của ChatGPT bao gồm đầu vào mã hóa, mô hình Transformer, giải mã và dự đoán, cùng với cơ chế học không giám sát Pretraining và Fine-tuning. Kiến trúc này cho phép ChatGPT xử lý thông tin ngôn ngữ và tạo ra phản hồi tự nhiên cho người dùng.

## **Câu 2: Những tiến bộ gần đây của Deep Learning và 20 bài báo về chủ đề ứng dụng deep learning trong xử lý ảnh y khoa có thời gian publish trong năm 2021-2022 và phân tích các phương pháp của từng bài báo**

Deep Learning là lĩnh vực nghiên cứu liên quan đến việc xây dựng và huấn luyện các mạng nơ-ron sâu để tự động học và rút trích kiến thức từ dữ liệu. Dưới đây là một số tiến bộ gần đây của Deep Learning:

1. GPT-3: Mô hình GPT-3 (Generative Pre-trained Transformer 3) là một trong những tiến bộ đáng chú ý. Với hơn 175 tỷ tham số, GPT-3 có khả năng sinh ra văn bản tự nhiên chất lượng cao và thực hiện nhiều nhiệm vụ ngôn ngữ phức tạp.

2. BERT: BERT (Bidirectional Encoder Representations from Transformers) là một mô hình Transformer huấn luyện trước (pretrained) cho việc rút trích đặc trưng từ dữ liệu ngôn ngữ tự nhiên. BERT đã đạt được kết quả ấn tượng trong nhiều tác vụ ngôn ngữ như phân loại văn bản và dịch máy.

3. EfficientNet: EfficientNet là một kiến trúc mạng nơ-ron chú trọng vào việc tối ưu hóa kích thước mô hình trong khi duy trì hiệu suất cao. Kiến trúc này đã đạt kết quả tốt trong nhiều bài toán như phân loại ảnh và nhận dạng vật thể.

4. GANs: GANs (Generative Adversarial Networks) là một mô hình Deep Learning bao gồm một mạng sinh (generator) và một mạng phân biệt (discriminator). GANs được sử dụng rộng rãi trong việc tạo ra ảnh, âm nhạc và văn bản tự nhiên tổng quát và chân thực.
