```python
import numpy as np
from collections import deque
class GraphMatrix:
  def __init__(self, v=None):
    self.v = v
    self.adj_list = np.full((self.v,self.v), 0)
  def count(self, i):
    return  self.adj_list[i].sum()

  def DFS(self, i):
    visited = [False] * self.v
    path = []

    def dfs_recursive(node):
        visited[node] = True
        path.append(node)

        for neighbor in range(self.v):
            if self.adj_list[node][neighbor] == 1 and not visited[neighbor]:
                dfs_recursive(neighbor)
    dfs_recursive(i)
    return path

  def BFS(self, i):
    visited = [False] * self.v
    path = []

    queue = deque()
    queue.append(i)
    visited[i] = True

    while queue:
        node = queue.popleft()
        path.append(node)

        for neighbor in range( self.v):
            if self.adj_list[node][neighbor] == 1 and not visited[neighbor]:
                queue.append(neighbor)
                visited[neighbor] = True

    return path

if __name__ == "__main__":
    v = 4
    a = [[0,1,0,0], [1,0,1,0], [0,1,0,1], [0,0,1,0]]
    g = GraphMatrix(4)
    for i in range(g.v):
      for j in range(g.v):
        g.adj_list[i,j] = a[i][j]
        print(g.adj_list[i,j], end=" ")
      print("\n")
    i=1
    print(f'Level of vertex {i}: {g.count(i)}')
    print(f'path from {i}: {g.DFS(i)}')

```
